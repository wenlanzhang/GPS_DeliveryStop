```python
import pandas as pd
import random
import string
import glob
import os
```

```python
Source = '/Users/zhangwenlan/Downloads/Tongji_RA/Data/Process/Processed/7Mannul/'
Destin = '/Users/zhangwenlan/Downloads/Tongji_RA/Data/Process/Processed/8QuantityCode/'
Order = '/Users/zhangwenlan/Downloads/Tongji_RA/Data/Process/Processed/FileID/'
OrderDestin = '/Users/zhangwenlan/Downloads/Tongji_RA/Data/Process/Processed/Code/'
```

```python
os.chdir(Source)

for file in glob.glob("*.csv"):
    df = pd.read_csv(file)
    dfOrder = pd.read_csv(Order + file)
    print(file)
    
    k = 0
    df['Quantity'] = 0.0
    ResultList = []

    
    for i, row in df.iterrows():
        if df.loc[i]['TrueStopFixed'] == 1:
            df.at[i, 'Quantity'] = dfOrder.loc[k]['TotalQuantity']
            
            letters = string.ascii_letters
            result_str = ''.join(random.choice(letters) for i in range(5)) + str(random.randint(0,50))
            df.at[i, 'Code'] = result_str
            dfOrder.at[k,'Code'] = result_str

            if result_str not in ResultList:
                ResultList.append(result_str)
            else:
                print('Error: File' + dfOrder.loc[i]['FileID'] + 'Stop' + dfOrder.loc[i]['StopID'])
                
            k += 1

    df.to_csv(Destin + file, index=False)
    dfOrder.to_csv(OrderDestin + file, index=False)
```
