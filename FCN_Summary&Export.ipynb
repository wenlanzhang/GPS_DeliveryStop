```python
import pandas as pd
import os
import glob
```

```python
# For Stops

Source = '/Users/zhangwenlan/Downloads/Tongji_RA/Data/Process/Processed/22/'

os.chdir(Source)
all_filenames = [i for i in glob.glob('*.csv')]

df_Combined = pd.concat([pd.read_csv(f) for f in all_filenames]) # Combine all files in the list
df_Combined = df_Combined.fillna(0)

df_TrueStops = df_Combined[df_Combined['TrueStopFixed'] == 1]
df_NonTrueStops = df_Combined[df_Combined['TrueStopFixed'] != 1]

Total = 'Sum.csv'
TrueStop = 'TrueStop.csv'
NonTrueStop = 'NonTrueStop.csv'

df_Combined.to_csv(Source + 'Processed/' + Total, index = False)
df_TrueStops.to_csv(Source + 'Processed/' + TrueStop, index = False)
df_NonTrueStops.to_csv(Source + 'Processed/' + NonTrueStop, index = False)

print(len(df_TrueStops))
```
