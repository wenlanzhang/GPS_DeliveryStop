按order data 选择前3天后1天的轨迹

```python
import pandas as pd
import glob
import os
```

```python
Source = '/Users/zhangwenlan/Downloads/Tongji_RA/Data/Process/Processed/24hFilter/'
Order = '/Users/zhangwenlan/Downloads/Tongji_RA/Data/Process/Processed/24hOrder/'
Destin = '/Users/zhangwenlan/Downloads/Tongji_RA/Data/Process/Processed/3_1/'
```

```python
os.chdir(Source)

TimeWindow1 = pd.to_timedelta('3 days 00:00:00')
TimeWindow2 = pd.to_timedelta('1 days 00:00:00')


for file in glob.glob("*.csv"):
    df = pd.read_csv(file)
    dfOrder = pd.read_csv(Order + file)
    print(file)
    
    del df['TrueStop']
    del df['CourseChangeTest']
    # Delete Columns

    
    dfOrder['ArrivalTime'] = pd.to_datetime(dfOrder['ArrivalTime'], format='%Y/%m/%d')
    df['StartTime'] = pd.to_datetime(df['StartTime'], format='%Y/%m/%d')
    
    MatchedTrajectory = list()
    
    for i, row in dfOrder.iterrows():
        TrajectoryStart = dfOrder.loc[i]['ArrivalTime'] - TimeWindow1
        TrajectoryEnd = dfOrder.loc[i]['ArrivalTime'] + TimeWindow2
        
        idx = df.index[(TrajectoryStart <= df['StartTime']) & (df['StartTime'] <= TrajectoryEnd)]
        for j in idx:
            if j not in MatchedTrajectory:
                MatchedTrajectory.append(j)

    df.iloc[MatchedTrajectory,:].to_csv(Destin + file, index=False)
    # print(MatchedTrajectory)
    # print(len(MatchedTrajectory))

    
```
